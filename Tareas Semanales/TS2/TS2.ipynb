# Primer Punto

Para calcular la ganancia pondre en funcion de Vo a las corrientes I1, I2 y I3

$VR2 = VC = -Vx$

$I1=-sCVx$

$I2=\frac{-Vx}{R2}$

$I3=\frac{-Vo}{R3}$

Se puede plantear que $Vx-I4R3=0$ entonces $I4=\frac{Vx}{R3}$. Luego $\frac{-I4}{sC}=Vy$, $I=5\frac{Vy}{R4}$ y por ultimo $Vo=-I5R4$

Juntando todas estas ecuaciones queda que $Vx=sCR3Vo$

Entonces las corrientes I1 y I2 quedan:

$I1=-s^2C^2R3Vx$

$I2=\frac{-sCR3Vo}{R2}$

Pongo IR1 en funcion de Vi y planteo corrientes en el nodo

$IR1=\frac{Vi}{R1}$

$IR1 = I1 + I2 + I3$

Si reemplazo las expresiones de corriente y despejo encuentro que la transferencia es:

$\frac{Vo}{Vi}=\frac{-R3}{R1}\frac{\frac{1}{C^2R3^2}}{s^2 + \frac{s}{CR2} + \frac{1}{C^2R3^2}}$

$wo^2=\frac{1}{C^2R3^2}$

$\frac{wo}{Q}=\frac{1}{CR2}$

$K=\frac{-R1}{R3}$

Por lo tanto la tranferencia termina quedando

$\frac{Vo}{Vi}=K\frac{wo^2}{s^2 + s\frac{wo}{Q} + wo^2}$

# Normalizacion

Para normalizar en frecuencia tomamos como norma wo y definimos una nueva variable $s'=\frac{s}{wo}$

Para normalizar en impedancia tomamos como norma R3 y quedan las impedancias nuevas:

$C'=CR3$ 

$R2'=\frac{R2}{R3}$ 

$R1'=\frac{R2}{R3}$

$R4'=\frac{R4}{R3}$

Juntando las dos normalizaciones queda la transferencia:

$\frac{Vo}{Vi}=\frac{1}{s'^2 + s' + 1}$

Entonces si seguimos con $R3=1$ 

$C=1$

Para cumplir la condicion de wo. Luego de la condicion de K vemos que $R1=1$
