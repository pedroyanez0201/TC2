Ahora realizamos el pasaje pasaaltos a pasabajos

$w_p = wp_n = 1$

$w_s = \frac{1}{ws_n} = 3.5$

Ya en pasabajos calculamos el epsilon

$\epsilon = \sqrt{10^\frac{alfa-max}{10}- 1} = 0.5$

Para el calculo de n planteamos

$alfa-min = 10*\log{(1 + \epsilon^2*(w_s^2)^n))}$

Luego de iterar vemos que la n mas chica que cumple con que el alfa minimo sea de 35 es n=4

Verificamos estos datos con spyder

```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt


# de la plantilla pasaaltos llegamos a la plantilla pasabajo
alfa_max = 1 # dB
alfa_min = 35
fs = 1000
fp = 3500

wp = fp * 2 * np.pi

ws = fs * 2 * np.pi/wp

wp_n= 1 #normalizada

# cuentas auxiliares

# epsilon 
ee = np.sqrt(10**(alfa_max/10)-1)

print ("epsilon", ee)

Om_p = 1
Om_s = 1/ws

for nn in range(2,6):
    alfa_min_w1 = 10*np.log10(1 + (ee**2) * (Om_s**2)**nn )
    print('con {:d} alfa minimo es {:f}'.format( nn, alfa_min_w1) )
```
